```python
import sys
import os
import requests
```

```python
song = """Куплет 1  
Снег на гильзах тает в ранней весне,  
А в щели окопа — белая астра.  
Пальцы сжимают ружье в полусне,  
Но сердце шепчет: «Вырастет к утру».  
Цветок не метафора, он — как заряд,  
Что рвёт пустоту между страхом и болью.  
Окопный закат, и в дыму винегрет,  
И в каждом вздохе — обещанье быть сильным.  

Куплет 2  
Семена в кармане, как письма из пламени,  
Они не сгорели, хоть дом был в огне.  
Снежинки на почву — молитва о памяти,  
А в мыслях: «Прорастут под весной в борьбе».  
Ты помнишь, как брат говорил про цветы?  
«Они отгоняют смерть, как заклинанье».  
Теперь ты — его корни, его черты,  
И каждый всход — это голос в отчаянье."""  
```

```python
input = {
#  "callback_url": None,
 "title": "Военная песня 1",
 "tags": "Гитара, военное настрокение",
 "prompt": song
}

headers = {
    'Content-Type': 'application/json',
    'Accept': 'application/json',
    'Authorization': f'Bearer {os.getenv("GEN_API")}'
}

url_endpoint = "https://api.gen-api.ru/api/v1/networks/suno"
response = requests.post(url_endpoint, json=input, headers=headers)
print(response.json())
```

```python
url_endpoint_answer = f"https://api.gen-api.ru/api/v1/request/get/{response.json()['request_id']}"
print(url_endpoint_answer)
```

```python
response_2 = requests.get(url_endpoint_answer, headers=headers)
print(response_2.json())
```
